# 01 – Data Loading and Exploration

This notebook loads raw motion data exported from Vision Mapper / MediaPipe (JSON) and performs basic visualization of joint trajectories over time.

## Steps

1. Configure paths to example JSON motion files.
2. Load one sample into pandas.
3. Inspect basic shapes and summary statistics.
4. Plot a few joint positions over time to visualize noise.


```python
import json
import pandas as pd
import matplotlib.pyplot as plt

SAMPLE_JSON_PATH = "D:/new_quant/data/raw/hand_wave.json"
```

```python
with open(SAMPLE_JSON_PATH, "r") as f:
    data = json.load(f)

df = pd.DataFrame(data)
print(df.head())
print(df.shape)

```

```python
wrist_x = [frame[0][0] for frame in df["landmarks"]]
#first index 0 is wrist, second index 0 is x position of wrist
plt.plot(df["timestamp"], wrist_x)
plt.xlabel("Timestamp (ms)")
plt.ylabel("Wrist X position")
plt.title("Wrist X over time")
plt.show()
```

```python
#first index is position of middle finger MCP, second index 0 is x position
palm_x = [frame[9][0] for frame in df['landmarks']]
palm_y = [frame[9][1] for frame in df['landmarks']]

plt.plot(palm_x, palm_y)
plt.xlabel("Palm x-position")
plt.ylabel("Palm y-position")
plt.title("Palm x pos over y pos")
plt.gca().invert_yaxis()
plt.show()
```
